```python
from sklearn.decomposition import PCA
pca = PCA(n_components= 50)
for i in range(len(pdmp1_ref)):
    temp = pdmp1-pdmp1_ref.iloc[i]
    temp = pca.fit_transform(temp)
    print(sum(pca.explained_variance_ratio_))
```

```python
def pipe(pdmp1,pdmp1_ref,num):
    pca = PCA(n_components= num)
    df_pdmp1 = pdmp1-pdmp1_ref.iloc[0]
    df_pdmp1 = pd.DataFrame(pca.fit_transform(df_pdmp1))
    for i in range(1,len(pdmp1_ref),1):
        temp = pdmp1-pdmp1_ref.iloc[i]
        temp = pd.DataFrame(pca.fit_transform(temp))
        df_pdmp1 = pd.concat([df_pdmp1,temp],axis=1)
    return df_pdmp1

number = 50
df_pdmp1 = pipe(pdmp1,pdmp1_ref,number)
df_pdmp2 = pipe(pdmp2,pdmp2_ref,number)
#df_pdmp3 = pipe(pdmp3,pdmp3_ref,number)
df_pdmp4 = pipe(pdmp4,pdmp4_ref,number)
df_pdmp5 = pipe(pdmp5,pdmp5_ref,number)
df_pdmp6 = pipe(pdmp6,pdmp6_ref,number)

df_pin1 = pipe(pin1,pin1_ref,number)
df_pin2 = pipe(pin2,pin2_ref,number)
#df_pin3 = pipe(pin3,pin3_ref,number)
df_pin4 = pipe(pin4,pin4_ref,number)
df_pin5 = pipe(pin5,pin5_ref,number)
df_pin6 = pipe(pin6,pin6_ref,number)

df_po1 = pipe(po1,po1_ref,number)
df_po2 = pipe(po2,po2_ref,number)
#df_po3 = pipe(po3,po3_ref,number)
df_po4 = pipe(po4,po4_ref,number)
df_po5 = pipe(po5,po5_ref,number)
df_po6 = pipe(po6,po6_ref,number)
```

```python
x= pd.DataFrame(x)
x.shape
```

```python
import sklearn

X_tr, X_te, y_tr, y_te = sklearn.model_selection.train_test_split(x, y_train_hot, test_size=0.2, random_state=42)
```

```python
model1 = tf.keras.models.Sequential([
    tf.keras.layers.Input(500,),
    tf.keras.layers.Dense(300, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(50, activation='relu'),
    tf.keras.layers.Dense(11, activation='softmax')
])

model1.compile(optimizer='adam',loss=tf.keras.losses.binary_crossentropy,metrics=['accuracy'])
```

```python
model1.summary()
```

```python
callbacks_list = [
    tf.keras.callbacks.ModelCheckpoint(
        filepath='best_model.{epoch:02d}-{val_loss:.2f}.h5',
        monitor='val_loss', save_best_only=True),
    tf.keras.callbacks.EarlyStopping(monitor='accuracy', patience=5)
]
```

```python
history = model1.fit(X_tr,y_tr,epochs=50,callbacks=callbacks_list,validation_data=(X_te,y_te))
```

```python
plt.figure(figsize=(6, 4))
plt.plot(history.history['accuracy'], 'r', label='Accuracy of training data')
plt.plot(history.history['val_accuracy'], 'b', label='Accuracy of validation data')
plt.plot(history.history['loss'], 'r--', label='Loss of training data')
plt.plot(history.history['val_loss'], 'b--', label='Loss of validation data')
plt.title('Model Accuracy and Loss')
plt.ylabel('Accuracy and Loss')
plt.xlabel('Training Epoch')
plt.ylim(0)
plt.legend()
plt.show()

```
